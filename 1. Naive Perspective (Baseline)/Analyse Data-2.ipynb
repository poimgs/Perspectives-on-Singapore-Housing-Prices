```
import pandas as pd
import os
```

# Import data

```
resale_flat_prices_file_path = os.path.join('Cleaned Data', 'resale flat prices.csv')
resale_flat_prices_df = pd.read_csv(resale_flat_prices_file_path)
```

# Modify data such that price per square meter is the predictor

```
def get_price_per_sqm(row):
    floor_area_sqm = row['floor_area_sqm']
    resale_price = row['resale_price']

    return int(resale_price / floor_area_sqm)

resale_flat_prices_df['resale_price_per_sqm'] = resale_flat_prices_df.apply(get_price_per_sqm, axis=1)
```

```
resale_flat_prices_df.drop(columns='resale_price', inplace=True)
```

# Train model

```
from pycaret.regression import *
```

```
setup(resale_flat_prices_df, 
      target='resale_price_per_sqm',
      session_id=42)
```

```
best = compare_models()
```

# Understand model

```
plot_model(best)
```

```
plot_model(best, save=True)
```

```
plot_model(best, plot='feature')
```

```
plot_model(best, plot='feature', save=True)
```

```
plot_model(best, plot='feature_all')
```

```
plot_model(best, plot='feature_all', save=True)
```

```
final_model = finalize_model(best)
```
